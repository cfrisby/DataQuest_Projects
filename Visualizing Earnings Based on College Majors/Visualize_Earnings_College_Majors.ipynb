# Visualizing Earnings by College Major

In this project, we will be explorting a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey. FiveThirtyEight cleaned the dataset and released in on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv')
```

```python
ax = recent_grads.plot("Sample_size", "Median", kind='scatter')
ax.set_title("Median vs. Sample_size")
```

From the above plot, there does not appear to be a strong correlation between the popularity of a major and the median salary associated with that major. The most popular majors appear to be average-below average in terms of median salary. Additionally, the very highest-paying and lowest-paying majors are near the bottom in number of students.

```python
ax = recent_grads.plot("Sample_size", "Unemployment_rate", kind='scatter', title='Unemployment_rate vs. Sample_size')
```

The above plot shows that the majors with the highest and lowest unemployment rates have relatively few college graduates. Meanwhile, majors with the greatest number of graduates seem to have average unemployment rates.

```python
ax = recent_grads.plot("Full_time", "Median", kind='scatter', title='Median vs. Full_time')
```

The above plot shows that there is not an obvious correlation between the number of full-time employees who graduated with a certain major, and the median salary of those individuals. In fact, this plot is very similar to the first scatter plot shown (Median vs. Sample_size). This makes sense, because majors with fewer graduates will in turn have fewer full-time employees.

```python
ax = recent_grads.plot("ShareWomen", "Unemployment_rate", kind='scatter', title='Unemployment_rate vs. ShareWomen')
```

From the above plot, there does not appear to be a relationship between unemployment rate and the share of women in a particular major.

```python
ax = recent_grads.plot("Men", "Median", kind='scatter', title='Median vs. Men')
```

```python
ax = recent_grads.plot("Women", "Median", kind='scatter', title='Median vs. Women')
```

The preceding two plots show median salary vs. the number of men and women respectively in each major. It is clear that many of the higher-paying majors have higher numbers of men than women. Additionally, many of the majors with large numbers of women have relatively low salaries.
